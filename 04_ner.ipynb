```python
def term_extraction(text,out,type):

  denotations=out['denotations']
  relavent_terms=[]

  if type=='drug':
    drug_terms=[]
    for i in denotations:
      if i['obj']== 'drug':
        drug_terms.append((i['span']['begin'],i['span']['end']))

    for i in drug_terms:
      start=i[0]
      end=i[1]
      relavent_terms.append(text[start:end])

  if type=='disease':
    species_terms=[]
    for i in denotations:
      if i['obj']== 'disease':
        species_terms.append((i['span']['begin'],i['span']['end']))

    for i in species_terms:
      start=i[0]
      end=i[1]
      relavent_terms.append(text[start:end])


  if type=='species':
    species_terms=[]
    for i in denotations:
      if i['obj']== 'species':
        species_terms.append((i['span']['begin'],i['span']['end']))

    for i in species_terms:
      start=i[0]
      end=i[1]
      relavent_terms.append(text[start:end])

  return(relavent_terms)
```

```python
import csv
import requests
import time

def query_plain(text, url="http://bern2.korea.ac.kr/plain"):
    return requests.post(url, json={'text': text}).json()

if __name__ == '__main__':
    csv_file = '/content/newbook55.csv'
    output_csv_file = '/content/output.csv'  # Path to the output CSV file

    with open(csv_file, 'r') as file:
        reader = csv.reader(file)
        next(reader)  # Skip header row

        with open(output_csv_file, 'w', newline='') as output_file:
            writer = csv.writer(output_file)
            writer.writerow(['Input', 'Annotations'])  # Write header row

            for row in reader:
                text = row[0]
                print("Input:", text)  # Print input line
                out = query_plain(text)
                print("Request Output:", out)  # Print request output

                annotations = []
                for annotation in out['annotations']:
                    mention = annotation['mention']
                    obj_type = annotation['obj']
                    if obj_type in ['drug', 'disease', 'chemical']:
                        annotations.append(f"{obj_type}: {mention}")

                annotations_str = ", ".join(annotations)
                print("Annotations:", annotations_str)
                writer.writerow([text, annotations_str])  # Write to the output CSV file

                # Introduce a delay to comply with the API limit
                time.sleep(0.33)
```

```python
import csv
import requests
import time

def query_plain(text, url="http://bern2.korea.ac.kr/plain"):
    return requests.post(url, json={'text': text}).json()

if __name__ == '__main__':
    csv_file = '/content/ph09.csv'
    output_csv_file = '/content/output.csv'  # Path to the output CSV file

    with open(csv_file, 'r') as file:
        reader = csv.reader(file)
        next(reader)  # Skip header row

        with open(output_csv_file, 'w', newline='') as output_file:
            writer = csv.writer(output_file)
            writer.writerow(['Input', 'Drugs', 'Diseases', 'Chemicals'])  # Write header row

            for row in reader:
                text = row[0]
                print("Input:", text)  # Print input line
                out = query_plain(text)
                print("Request Output:", out)  # Print request output

                drugs = []
                diseases = []
                chemicals = []
                for annotation in out['annotations']:
                    mention = annotation['mention']
                    obj_type = annotation['obj']
                    if obj_type == 'drug':
                        drugs.append(mention)
                    elif obj_type == 'disease':
                        diseases.append(mention)
                    elif obj_type == 'chemical':
                        chemicals.append(mention)

                writer.writerow([text, ', '.join(drugs), ', '.join(diseases), ', '.join(chemicals)])  # Write to the output CSV file

                # Introduce a delay to comply with the API limit
                time.sleep(0.33)
```
